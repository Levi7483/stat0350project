```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
```

```python
files = os.listdir('/Users/levimunster/Desktop/VS_Code/MiddleburyCS/stat0350/steam_db_playerdata')
```

```python
achievement_df = pd.DataFrame()
```

```python
for file in files:
    info = file.split('_')    
    game = info[0]
    id = info[1][:-4]

    URL = f"https://steamcommunity.com/stats/{id}/achievements"

    html = requests.get(URL, timeout=10).text
    soup = BeautifulSoup(html, "html.parser")
    counter = 4

    row = {'game': file[:-4]}
    counter = 4
    while True:
        percent = soup.select_one(
            f"#mainContents > div:nth-child({counter}) > div.achieveTxtHolder > div.achievePercent"
        )
        if percent is None:
            break
        row[f"achievement_{counter-3}"] = float(percent.text[:-1])
        counter += 1
    print(row)
    achievement_df = pd.concat(
    [achievement_df, pd.DataFrame([row])],
    ignore_index=True
    )
```
