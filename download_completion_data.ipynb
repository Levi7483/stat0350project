```python
skip_list = []
for game in skipped_list:
    print(f'Starting on {game}')
    search.clear()
    search.send_keys(game)
    search.send_keys(Keys.ENTER)
    print(driver.current_url)

    game_link_locator = (By.LINK_TEXT, game)
    time.sleep(2)
    if "No Results Found, Try A New Search" in driver.page_source:
        print(f'No results found. Skipping {game}')
        skip_list.append(game)
        continue
    
    try:
        max_retries = 3
        for attempt in range(max_retries):
            try:
                print(f'Searching for {game_link_locator}')
                # Re-locate AND click within the try block every single time
                WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable(game_link_locator)
                ).click()
                print(f"Click successful on attempt {attempt + 1}")
                break # Exit the loop if successful

            except StaleElementReferenceException:
                print(f"Stale element encountered on attempt {attempt + 1}. Retrying...")
                if attempt == max_retries - 1:
                    raise # If it's the last attempt and still failing, re-raise the error
                time.sleep(0.5) # Optional: add a small pause before retrying
            except TimeoutError:
                break
        print('Searching for data...')
        main_story = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "(//ul/li[h4 and h5])[1]/h5"))
            ).text
        main_sides = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "(//ul/li[h4 and h5])[2]/h5"))
            ).text
        completionist = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "(//ul/li[h4 and h5])[3]/h5"))
            ).text
        print(f'Found {main_story},{main_sides},{completionist} for {game}')
        print(driver.current_url)
        row = [game, main_story[:-5], main_sides[:-5], completionist[:-5]]
        print(row)
        output.loc[len(output)] = row
        print(f'Done with {game}')
    except Exception as e:
        print(e)
        skip_list.append(game)
        
```
