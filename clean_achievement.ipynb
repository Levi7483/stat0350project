```python
rows = []
achievement_cols = [c for c in achievement_df.columns if c.startswith("achievement_")]

for row in achievement_df.itertuples(index=False):
    values = np.array([getattr(row, c) for c in achievement_cols], dtype=float)
    values = values[~np.isnan(values)]

    num_achievements = len(values)
    if num_achievements < 4:
        continue

    early_end = num_achievements // 3
    mid_end = 2 * num_achievements // 3

    early_avg = values[:early_end].mean()
    mid_avg   = values[early_end:mid_end].mean()
    late_avg  = values[mid_end:].mean()

    rows.append({
        "game": row.game,
        "early_percentage": early_avg,
        "mid_percentage": mid_avg,
        "late_percentage": late_avg,
        "num_achievements": num_achievements
    })

output_df = pd.DataFrame(rows)
```
